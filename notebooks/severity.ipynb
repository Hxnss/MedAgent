```python
import pandas as pd

info = pd.read_excel("dataset.xlsx", sheet_name="Info Penyakit")

info_penyakit = info.set_index("Nama Penyakit").to_dict("index")

print(info_penyakit["Malaria"])
```

```python
aturan_tindakan = {
    "HIGH": {
        "warna": "🔴 Merah",
        "tindakan": "SEGERA ke IGD / UGD rumah sakit terdekat.",
        "urgensi": "Darurat — butuh penanganan medis segera."
    },
    "MODERATE": {
        "warna": "🟡 Kuning",
        "tindakan": "Periksa ke dokter dalam 24 jam.",
        "urgensi": "Perlu perhatian — jangan ditunda lama."
    },
    "LOW": {
        "warna": "🟢 Hijau",
        "tindakan": "Istirahat di rumah & pantau gejala. Ke dokter bila memburuk.",
        "urgensi": "Ringan — umumnya bisa dirawat di rumah."
    }
}

print(aturan_tindakan["HIGH"])
```

```python
def ambil_info_penyakit(nama_penyakit):
    # Ambil data dari sheet Info Penyakit
    data = info_penyakit.get(nama_penyakit)

    if data is None:
        return {"error": f"Info untuk '{nama_penyakit}' tidak ditemukan."}

    severity = data["Risk Severity"]            # HIGH / MODERATE / LOW
    tindakan = aturan_tindakan.get(severity, {})

    return {
        "penyakit": nama_penyakit,
        "kategori": data["Kategori"],
        "deskripsi": data["Deskripsi"],
        "dokter": data["Dokter Spesialis"],
        "severity": severity,
        "warna": tindakan.get("warna"),
        "urgensi": tindakan.get("urgensi"),
        "tindakan": tindakan.get("tindakan"),
    }

# Uji untuk beberapa penyakit
for p in ["Malaria", "GERD", "Stroke"]:
    hasil = ambil_info_penyakit(p)
    print(f"\n=== {hasil['penyakit']} ({hasil['severity']}) {hasil['warna']} ===")
    print("Tindakan :", hasil["tindakan"])
    print("Dokter   :", hasil["dokter"])
    print("Deskripsi:", hasil["deskripsi"][:70], "...")
```
